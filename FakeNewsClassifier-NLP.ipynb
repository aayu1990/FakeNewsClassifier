```python
# removing hyper links from the text
df["text_no_link"] = df['text'].apply(lambda x: re.split('http:\/\/.*', str(x))[0])
df.head()
```

```python
df["text_no_link"][0]
```

```python
df = df.drop(["text"], axis=1)
df.head()
```

```python
# Remove stopwords and remove words with 2 or less characters
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
# Apply the function to the dataframe
df['clean'] = df['text_no_link'].apply(preprocess)
df.head()
```

```python
df = df.drop(["text_no_link"], axis=1)
df.head()
```

```python
# Obtain the total words present in the dataset
list_of_words = []
for i in df.clean:
    for j in i:
        list_of_words.append(j)
```

```python
list_of_words
```

```python
len(list_of_words)
```

```python
# Obtain the total number of unique words
total_words = len(list(set(list_of_words)))
total_words
```

```python
# join the words into a string
df['clean_joined'] = df['clean'].apply(lambda x: " ".join(x))
df
```

```python
df = df.drop(["clean"], axis=1)
df
```

```python
# plot the word cloud for text that is Real
plt.figure(figsize = (20,20)) 
wc = WordCloud(max_words = 2000 , width = 1600 , height = 800 , stopwords = stop_words).generate(" ".join(df[df.Rating == 1].clean_joined))
plt.imshow(wc, interpolation = 'bilinear')
```

```python
# plot the word cloud for text that is Fake
plt.figure(figsize = (20,20)) 
wc = WordCloud(max_words = 2000 , width = 1600 , height = 800 , stopwords = stop_words).generate(" ".join(df[df.Rating == 0].clean_joined))
plt.imshow(wc, interpolation = 'bilinear')
```
