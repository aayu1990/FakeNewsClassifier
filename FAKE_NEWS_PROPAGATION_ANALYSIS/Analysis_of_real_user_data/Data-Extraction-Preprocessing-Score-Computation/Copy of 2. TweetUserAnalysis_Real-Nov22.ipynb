### User Followers to Following ratio

```python
def ratio(df):
    if(df['following_count'] == 0):
        df['followers_following_ratio'] = 0
    else:
        df['followers_following_ratio'] = df['followers_count']/df['following_count']
    return df
```

```python
dfFF_Ratio = df.apply(ratio,axis=1)
```

```python
dfFF_Ratio
```

```python
sorted_dfFF_Ratio = dfFF_Ratio.sort_values(by=['followers_following_ratio','user_id'], ascending=False)
sorted_dfFF_Ratio
```

## Status to followers ratio

```python
def ratioSF(df):
    if(df['followers_count'] == 0):
        df['status_followers_ratio'] = 0
    else:
        df['status_followers_ratio'] = df['status_count']/df['followers_count']
    return df
```

```python
dfSF_Ratio = df.apply(ratioSF,axis=1)
```

```python
sorted_dfSF_Ratio = dfSF_Ratio.sort_values(by=['status_followers_ratio','user_id'], ascending=False)
sorted_dfSF_Ratio
```

```python
from datetime import date

current_date = date.today() 
print("Current year:", type(current_date.year))
```

```python
def dateFormatingtoYear(df):
    df['created_year'] = datetime.strptime(df['created_at'], '%a %b %d %H:%M:%S +0000 %Y').strftime('%Y')
    df['active_years'] = date.today().year - int(df['created_year'])
    return df
```

```python
df_Year = df.apply(dateFormatingtoYear,axis=1)
```

```python
df_Year
```

```python
sorted_dfActiveYears = df_Year.sort_values(by=['active_years','user_id'], ascending=False)
sorted_dfActiveYears
```
