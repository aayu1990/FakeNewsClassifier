```python
import os as os
import glob
import json
import pandas as pd
```

```python
totalContents = []
json_dir_name = '/Users/priyankaraju/Documents/Jupyter Notebooks/Sept2020/true-news/'
```

```python
for file_name in [file for file in os.listdir(json_dir_name) if file.endswith('.json')]:
  with open(json_dir_name + file_name) as json_file:
    data = json.load(json_file)
    totalContents.append(data)
```

```python
print(len(totalContents))
```

```python
df = pd.DataFrame(totalContents)
```
