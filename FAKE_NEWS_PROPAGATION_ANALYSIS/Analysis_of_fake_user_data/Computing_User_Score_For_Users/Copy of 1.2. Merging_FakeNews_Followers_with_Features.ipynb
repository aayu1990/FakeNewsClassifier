Computing Followers to following ratio


```
def ratio(df):
    if(df['foll_following'] == 0):
        df['followers_following_ratio'] = 0
    else:
        df['followers_following_ratio'] = df['foll_follower']/df['foll_following']
    return df
```

```
dfFF_Ratio = df.apply(ratio,axis=1)
dfFF_Ratio
```

```
dfFF_Ratio= dfFF_Ratio['followers_following_ratio']
dfFF_Ratio
```
