Computing Active Years


```
def dateFormatingtoYear(df):
  df['foll_profileTime'] = datetime.strptime(df['foll_profileTime'], '%Y-%m-%d %H:%M:%S').strftime('%Y')
  df['active_years'] = date.today().year - int(df['foll_profileTime'])
  return df
```

```
def verified_score(df):
    if(df['foll_isVerified'] == 1.0):
        df['verified_score'] = 0.35
    else:
        df['verified_score'] = 0.00
    
    if(df['active_years'] >= 0 and df['active_years'] <= 5.0):
        df['active_year_score'] = 0.05
    elif(df['active_years'] > 5.0 and df['active_years'] <= 8.0):
        df['active_year_score'] = 0.1
    elif(df['active_years'] > 8.0 and df['active_years'] <= 10.0):
        df['active_year_score'] = 0.15
    else:
        df['active_year_score'] = 0.2
        
    if(df['followers_following_ratio'] >= 0 and df['followers_following_ratio'] <= 0.2141647221627625):
        df['ff_score'] = 0.05
    elif(df['followers_following_ratio'] > 0.2141647221627625 and df['followers_following_ratio'] <= 0.6252412059741004):
        df['ff_score'] = 0.1
    elif(df['followers_following_ratio'] > 0.6252412059741004 and df['followers_following_ratio'] <= 1.0026946251574234):
        df['ff_score'] = 0.15
    else:
        df['ff_score'] = 0.2
        
    if(df['foll_favouritesCount'] >= 0 and df['foll_favouritesCount'] <= 130.75):
        df['favorites_count_score'] = 0.05
    elif(df['foll_favouritesCount'] > 130.75 and df['foll_favouritesCount'] <= 2401.0):
        df['favorites_count_score'] = 0.1
    elif(df['foll_favouritesCount'] > 2401.0 and df['foll_favouritesCount'] <= 18424.25):
        df['favorites_count_score'] = 0.15
    else:
        df['favorites_count_score'] = 0.2
        
    if(df['status_followers_ratio'] >= 0 and df['status_followers_ratio'] <= 0.6637408396243003):
        df['sf_score'] = 0.05
    elif(df['status_followers_ratio'] > 0.6637408396243003 and df['status_followers_ratio'] <= 2.6814854817064897):
        df['sf_score'] = 0.1
    elif(df['status_followers_ratio'] > 2.6814854817064897 and df['status_followers_ratio'] <= 9.769490644490645):
        df['sf_score'] = 0.15
    else:
        df['sf_score'] = 0.2
        
    df['followers_weights'] = df['verified_score'] + df['active_year_score'] + df['ff_score'] +  df['favorites_count_score'] + df['sf_score']
    df['final_weights'] = df['weights'] + df['followers_weights']
    return df
```
