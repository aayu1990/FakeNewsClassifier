1. Iterate through the user
2. Find the number of rows for each user
3. Sort the rows based on weights
4. Take the top 5% of the total rows


```
user_id_list = list(df['user_id'].unique())
user_id_list
len(user_id_list)
```

```
test = df[df['user_id']== 1084.0].sort_values('final_weights',ascending=False)
##test.reset_index()
len(test)
round(len(test)*0.20)
```

```
result = []
for user_id in user_id_list:
  user_df = df[df['user_id']== user_id].sort_values('final_weights',ascending=False)
  user_df.reset_index()
  df_length = round(len(user_df)*0.25)
  user_df = user_df[:df_length]
  result.append(user_df)
result_df = pd.concat(result)
result_df = result_df.reset_index(drop=True)
  
```
