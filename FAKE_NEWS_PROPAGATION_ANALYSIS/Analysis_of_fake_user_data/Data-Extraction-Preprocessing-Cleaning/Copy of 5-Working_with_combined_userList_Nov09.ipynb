```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('combined_user_followerList_Nov09.csv')
df = df.head(100)
```

```python
df.columns
```

```python
new_df = pd.DataFrame(columns = ['user_id', 'weights', 'followers'])
```

```python
for index, row in df.iterrows():
    followerCount = len(row['followers'])
    for i in ast.literal_eval(row['followers']):
#         print(i)
        new_df = new_df.append({'user_id':row['user_id'],'weights':row['weights'],'followers':i },ignore_index = True)
```

```python
new_df.to_csv('user-follower-weighted_100_Nov09.csv',index=False)
```
