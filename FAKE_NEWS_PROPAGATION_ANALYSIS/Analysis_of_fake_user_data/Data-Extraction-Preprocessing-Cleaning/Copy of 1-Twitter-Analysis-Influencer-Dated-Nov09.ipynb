```python
import os
import glob
import json
import pandas as pd
```

```python
df = pd.read_csv('finalTweets.csv')
```

```python
df.head(5)
```

```python
df = df[df['user'].notna()]
```

```python
df
```

```python
df_userdetails = pd.DataFrame(df['user'])
df_userdetails
```

```python
userIdArr = []
userFollowerArr = []
userFollowingArr = []
userStatusArr = []
userCreatedAtArr = []
userFavoriteCntArr = []
userVerifiedArr = []
```

```python
for i, j in df_userdetails.iterrows(): 
#     userDictArry.append(eval(j))
    user = eval(j['user'])
#     print(i)
    userIdArr.append(user['id'])
    userFollowerArr.append(user['followers_count'])
    userFollowingArr.append(user['friends_count'])
    userStatusArr.append(user['statuses_count'])
    userCreatedAtArr.append(user['created_at'])
    userFavoriteCntArr.append(user['favourites_count'])
    userVerifiedArr.append(user['verified'])
```

```python
print(len(userIdArr),len(userFollowerArr),len(userFollowingArr),len(userStatusArr),len(userCreatedAtArr))
```

```python
twitter_list = [userIdArr,userFollowerArr,userFollowingArr,userStatusArr,userCreatedAtArr,userFavoriteCntArr,userVerifiedArr]

new_df = pd.DataFrame (twitter_list).transpose()
new_df.columns = ['user_id','followers_count','following_count','status_count','created_at','favorites_count','verified']
print (new_df)
```

```python
new_df.to_csv('FakeNews_Data_Nov09.csv',index=False)
```
